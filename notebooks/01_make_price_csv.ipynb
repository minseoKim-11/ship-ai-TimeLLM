```python
# notebooks/01_prices_build.ipynb
from pykrx import stock
import pandas as pd, os

def save_prices(ticker, name):
    df = stock.get_market_ohlcv_by_date("20230101","20251231", ticker).reset_index()
    df = df.rename(columns={"날짜":"date","시가":"open","고가":"high","저가":"low","종가":"close","거래량":"volume"})
    df["market"]="KOSPI"
    os.makedirs("/workspace/ship-ai/data/silver/prices_daily", exist_ok=True)
    df.to_csv(f"/workspace/ship-ai/data/silver/prices_daily/{name}.csv", index=False)

save_prices("329180","HD현대중공업")
save_prices("010140","삼성중공업")
save_prices("042660","한화오션")
```
